## 1. Data Exploration

```python
import numpy as np
import pandas as pd
import scipy

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#import data

df = pd.read_csv('data/clean/segmentation data.csv')
```

```python
df = df.drop('ID', axis=1)
df.head()
```

```python
#correlation 
df.corr()
```

```python
#Correlation between different variables
corr_df = df.corr()

f, ax = plt.subplots(figsize=(15,10))

#Configure custom diverging colormap
cmap = sns.diverging_palette(230,20, as_cmap=True)

#Draw map
sns.heatmap(corr_df, annot=True, cmap=cmap)
```

```python
#visualize raw data

fig, ax = plt.subplots(figsize = (12, 8))

x = df['Age']
y = df['Income']
ax.scatter(x, y)
ax.set(title='Income vs Age', ylabel='Income', xlabel='Age')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#feature standardization so that all features have equal weight for modelling.
# Otherwise, in our case Income would be considered much more important than Education for Instance. 
# We do not know if this is the case, so we would not like to introduce bias.
scaler = StandardScaler()
segmentation_std = scaler.fit_transform(df)
```
